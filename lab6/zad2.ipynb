```julia
import Pkg
Pkg.add("Plots")
Pkg.add("Roots")
Pkg.add("ForwardDiff")
using Plots
using Roots
using ForwardDiff
```

```julia
f(x)=exp(x) - 1/(10*x)^2
plot(f, 0.01,0.25)
```

```julia
fpy =[]
fpx =[]
a=0.08
b=0.15
function make_verbose(f, fpy, fpx, a, b)
    function wrapFp(x)
        push!(fpx, x)
        push!(fpy, f(x))
        f(x)
    end
    find_zero(wrapFp, (a, b), FalsePosition(12))
end
```

```julia
make_verbose(f, fpy, fpx, a, b)
scatter(fpx, fpy)
```

```julia
npy =[]
npx =[]
D(f) = x ->ForwardDiff.derivative(f, float(x))
a=0.01
b=0.25
function make_verbose_newton(f, npy, npx, a, b)
    dfx = []
    dfy =[]
    function wrap_np(x)
        push!(npx, x)
        push!(npy, f(x))
        f(x)
    end
    
    function D(x)
        Dp(f) = xl->ForwardDiff.derivative(f, float(xl))
        push!(dfy, Dp(f)(x))
        push!(dfx, Dp(x))
        Dp(f)(x)
    end
    find_zero((wrap_np, D), a, Roots.Newton())
end
```

```julia
make_verbose_newton(f, npy, npx, a,b)
scatter(npx, npy)
```

```julia
ssx=[]
ssy=[]
function make_verbose_ss(f, ssx, ssy, a, b)
    function wrap_ss(x)
        push!(ssx, x)
        push!(ssy, f(x))
        f(x)
    end
    find_zero(wrap_ss, a, Order2(), verbose=true)
end
make_verbose_ss(f, ssx, ssy, a, b)
scatter(ssx, ssy)
```
